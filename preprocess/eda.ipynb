```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('./data/food_data.csv')
print(f'data size: {len(data)}')
data.head()
```

```python
data_group_by_large= data.groupby('식품대분류')['식품명'].apply(list)
print(len(data['식품대분류'].unique()))
# print(data['식품대분류'].unique())
data_group_by_large
```

```python
data_group_by_large['len']= data_group_by_large.apply(lambda x: len(x))
print(data_group_by_large['len'])
print(f"max: {data_group_by_large['len'].max()}, min: {data_group_by_large['len'].min()}, mean: {data_group_by_large['len'].mean()}")
```

```python
tmp= data_group_by_large['len'].sort_values(ascending= False)
print(tmp)

plt.barh(tmp.index, tmp.values)
```

```python
data_group_by_small= data.groupby('식품상세분류')['식품명'].apply(list)
print(len(data['식품상세분류'].unique()))
# print(data['식품상세분류'].unique())
data_group_by_small
```

```python
data_group_by_small['len']= data_group_by_small.apply(lambda x: len(x))
print(data_group_by_small['len'])
print(f"max: {data_group_by_small['len'].max()}, min: {data_group_by_small['len'].min()}, mean: {data_group_by_small['len'].mean()}")

for i in range(len(data_group_by_small)):
    # print(type(data_group_small['len'][i]))
    if data_group_by_small['len'][i]== data_group_by_small['len'].max():
        print(data_group_by_small.index[i])
        break;
```

```python
small_list= data['식품상세분류'].unique()
large_list= data['식품대분류'].unique()

cnt= 0
for s in small_list:
    if s in large_list:
        cnt+=1
print(f'duplicated count : {cnt}')
```
