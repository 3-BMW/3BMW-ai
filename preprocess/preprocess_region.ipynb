```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import json
from tqdm import tqdm
```

```python
DATA_PATH= '../data/food_data.csv'
JSON_PATH= '../data/region_major_set_data.json'

data= pd.read_csv(DATA_PATH)
# data.head()
print(len(list(set(data[data['상용제품']=='품목대표']['식품명'].values))))
print(data)
```

```python
# 지역은 짧으니까..?
region_list= [ 
'충주', '춘천', '광양', '고흥', '수원', '기장', '태안', '삼척', '장흥', '괴산', 
'대전', '포항', '속초', '강릉', '평창', '정선', '인천', '여수', '안성', '밀양', 
'거제', '영주', '전주', '산청', '부안', '영동', '광주', '마산', '안동', '영암', 
'나주', '순천', '남원', '고성', '부산', '강화', '사천', '동해', '제주', '완도', 
'진주', '진도', '군산', '고창', '서해', '경기', '문경', '청평', '울산', '영덕', 
'대구', '안산', '신안', '목포', '남해', '양양', '진해', '하동', '영광', '서천', 
'독도', '충남', '서주', '의정부', '강화도', '구룡포', '주문진']

print(sorted(data['지역 / 제조사'].unique(), key= lambda x: len(x))[:300])
# data_group_by_small= data.groupby('지역 / 제조사')['식품명'].apply(list)
# data_group_by_small.values[:10]
```

```python
from tqdm import tqdm 

region_df= pd.DataFrame(columns= ['식품명', '지역'])
for i in tqdm(range(len(data))):
    if data.iloc[i]['지역 / 제조사'] in region_list:
        region_df= region_df.append({'식품명': data.iloc[i]['식품명'], '지역': data.iloc[i]['지역 / 제조사']}, ignore_index= True) 
```

```python
region_df.head()
```

```python
region= region_df.groupby('지역')['식품명'].apply(set).apply(list)
region
```

```python
region_data= []
for idx in tqdm(range(len(region.index))):
    for item in region.values[idx]:
        tmp= [region.index[idx], item]
        region_data.append(tmp)
print(len(region_data))
print(region_data[-100:])



    
```

```python
with open (JSON_PATH, 'w') as f:
    f.write(json.dumps(region_data, ensure_ascii=False))
```
